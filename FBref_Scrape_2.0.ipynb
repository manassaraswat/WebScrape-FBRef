```python
def fbref_clearner(df):
    temp = df.columns
    cols = []
    index_columns = []
    for item in temp:
        if item[0].split(':',1)[0] == 'Unnamed':
            index_columns.append(item[1])
        else:
            cols.append(item[0] + '_' + item[1])
    index_columns = index_columns[:-1]
    total_cols = index_columns + cols
    df = df.droplevel(0,axis=1)
    df.columns = total_cols
    temp_df = df.loc[:, cols].apply(pd.to_numeric, errors='coerce')
    temp_df.fillna(0, inplace=True)
    master_df = pd.concat([df[['Player','Nation','Pos','Squad','Comp','Age','Born']], temp_df], axis = 1)
    return master_df

```

```python
# def fbref_dropheader(df):
#     my_df = df.copy()
#     my_df = my_df.droplevel(0,axis=1)
#     return(my_df)

def fbref_cleaner(df):
    all_columns = list(df.columns)
    unwanted_columns = {'Rk', 'Player','Nation','Pos','Squad','Comp','Age','Born','Matches'}
    index_columns = [ele for ele in all_columns if ele not in unwanted_columns]
    temp_df = df.loc[:, index_columns].apply(pd.to_numeric, errors='coerce')
    temp_df.fillna(0, inplace=True)
    master_df = pd.concat([df[['Player','Nation','Pos','Squad','Comp','Age','Born']], temp_df], axis = 1)
    return master_df
```
